```python
#importing the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
```

```python
#creating a readable file for the observations.csv file
with open('/home/essetu/Documents/Datasets/Codecademy projects/biodiversity-starter/biodiversity_starter/observations.csv') as observations_file:
    observations_read = pd.read_csv(observations_file)
```

```python
#creating a readable file for the species.csv file
with open('/home/essetu/Documents/Datasets/Codecademy projects/biodiversity-starter/biodiversity_starter/species_info.csv') as species_file:
    species_read = pd.read_csv(species_file)
```

```python
#Creating a larger file that includes all the information available using pd.merge()
class_observations = pd.merge(observations_read, species_read, on= 'scientific_name', how= 'outer')

#Dropping duplicates with the .drop_duplicates function. The input 'inplace=True' means that the function doesn't
#create a new file but rather changes the existing file.
class_observations.drop_duplicates(inplace=True, subset=['scientific_name', 'park_name', 'observations'],
                                   keep='first')

#Renaming common_names to be more coherent with the other  labels.The same thing applies with 'inplace=True'.
class_observations.rename(inplace=True, columns={'common_names' : 'common_name'})

#Replace NaN with low risk
class_observations.fillna('Low Risk', inplace=True)

#Create a sub-file with only recovering species

life_forms_recovering = class_observations[class_observations['conservation_status'] == 'In Recovery']
life_forms_recovering.head()
life_forms_recovering.tail()
```

```python
#Truncating the common_name values because they are often too long to fit in a graph
class_observations['trunc_common_name'] = (class_observations['common_name'].str.split().str[-2:].str.join(' '))
```

```python
#Creating a sub-file for each category of life forms.
for category, group in class_observations.groupby('category'):
     group.to_csv(f'{category}.csv', index=False)
        
#Creating a readable file for mammals within the larger file
mammals_observations = pd.read_csv('Mammal.csv')

#Creating a sub-file for mammals in each national park
for (park_name), group in mammals_observations.groupby('park_name'):
     group.to_csv(f'{park_name}.csv', index=False)
    
#Creating a readable file for mammal observations in Yosemite
yosemite_mammals_observations = pd.read_csv('Yosemite National Park.csv')

#Creating a sub-file for each conservation status of mammals in yosemite (this may be useless, but it allows me some 
#flexibility that I like.
yosemite_mammals_status = {}

for status,  _ in yosemite_mammals_observations.groupby('conservation_status'):
    yosemite_mammals_status[status] = pd.read_csv(f'{conservation_status}.csv')

    #Truncating the common name for the recovering populations because the name was entered in a different format, sadly.
if 'In Recovery' in yosemite_mammals_status:
    yosemite_mammals_status['In Recovery']['trunc_common_name'] = (yosemite_mammals_status['In Recovery']['common_names'].str.split().str[:2].str.join(' '))

print(yosemite_mammals_status['In Recovery'])
```

```python
#Using pd.concat() to join the yosemite_mammals dataframes into one
yosemite_mammals_at_risk = pd.concat([yosemite_mammals_recovering, yosemite_mammals_concern, yosemite_mammals_threatened, 
                                      yosemite_mammals_endangered])
 
#Setting the figure size before I create a plot
plt.figure(figsize=(10,6))

#Creating a color palette for the hue portion of my graph
vulnerability_palette = {'Low Risk' : 'forestgreen', 'In Recovery' : 'gold', 'Species of Concern': 'darkorange', 
                         'Threatened' : 'orangered', 'Endangered' : 'red'}

#Creating a barplot with seaborn including the observations, common name, and conservation status
sns.barplot(yosemite_mammals_at_risk, x='trunc_common_name', y = 'observations',
            hue = 'conservation_status', palette = vulnerability_palette )

#Rotating the names on the x axis so they are more legible
plt.xticks(rotation = 90)

#Changing the label names with plt.xlabel and plt.ylabel
plt.xlabel('Common Names')
plt.ylabel('Population Observed - Number of Animals')

#Adding a title with plt.suptitle and a subtitle with plt.title
plt.suptitle('Comparing Mammal Population Numbers in Yosemite National Park')
plt.title('Part 1: Vulnerable Species' )

#Changing the legend title with plt.legend
plt.legend(title='Conservation Status')
```
